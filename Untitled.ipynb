```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import time
```

```python
data = np.loadtxt('319bec_avg.txt')
```

```python
def add_noise(data):
    x,y = data.shape
    add = np.random.randn(x,y)/20
    return data + add
```

```python
while(1 > 0):
    time.sleep(10)
    np.savetxt('newimage.txt',add_noise(data))
```

```python
plt.imshow(add_noise(data))
```

```python
import matplotlib.animation as anim

def plot_cont(fun, xmax):
    y = []
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)

    def update(i):
        yi = fun()
        y.append(yi)
        x = range(len(y))
        ax.clear()
        ax.plot(x, y)
        print(i, ': ', yi)

    a = anim.FuncAnimation(fig, update, frames=xmax, repeat=False)
    plt.show()
```
